IMPORT LIBRARIES

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Load The Dataset

```python
df = pd.read_csv("D://Data Science//netflix1.csv")
df.head()
```

Clean The Dataset

```python
df = df.dropna(subset=["show_id","type","title","director","country","date_added","rating","release_year","duration","listed_in"])

type_counts = df["type"].value_counts()

df
```

Creating Bar Chart

```python

plt.figure(figsize=[6,4])

plt.bar(type_counts.index , type_counts.values , color=["skyblue" , "orange"])

plt.title("Number Of Movies VS TV Shows On Netflix")

plt.xlabel("Type")
plt.ylabel("Counts")

plt.tight_layout

plt.savefig("movies_vs_tvshows.png")

plt.show()
```

Creating Pie Chart

```python
import matplotlib.pyplot as plt

rating_counts = df["rating"].value_counts()  # Note the parentheses

plt.figure(figsize=[8,6])
plt.pie(
    rating_counts, 
    labels=rating_counts.index, 
    startangle=90, 
    autopct="%1.1f%%",  # Show percentages with 1 decimal
    shadow=True,        # Optional: adds a shadow
    explode=[0.05]*len(rating_counts)  # Optional: separates slices a bit
)
plt.title("Percentage of Content Ratings")
plt.tight_layout()  # Correctly call the function
plt.show()
```

Creating Histogram

```python
movie_df = df[df["type"] == "Movie"].copy()
movie_df["duration_int"] = movie_df["duration"].str.replace('min' , '').astype(int)

plt.figure(figsize=[8,6])

plt.hist(movie_df["duration_int"] , bins = 30 , color="purple" , edgecolor = "black")

plt.title("Distribution Of Movie Duration")

plt.xlabel("Duration (mints)")
plt.ylabel("Number of Movies")

plt.tight_layout()

plt.savefig("movie_duration_histogram.png")

plt.show()
```

Creating Scatter Plot

```python
release_counts = df["release_year"].value_counts()

plt.figure(figsize=[10,6])

plt.scatter(release_counts.index , release_counts.values , color = "red")

plt.title("Release Number Of Shows")

plt.xlabel("Release Years")
plt.ylabel("Number Of Shows")

plt.tight_layout()

plt.savefig("release_year_scatter.png")

plt.show()
```

Creating Horizantal Bar Chart

```python
country_counts = df["country"].value_counts().head(10)

plt.figure(figsize=[8,6])

plt.barh(country_counts.index , country_counts.values , color = "teal")

plt.title("Top 10 Countries By Number Of Shows")

plt.xlabel("Number Of Shows")
plt.ylabel("Country")

plt.tight_layout()

plt.savefig("top-10_countries.png")

plt.show()
```

Creating Subplots

```python
import pandas as pd
import matplotlib.pyplot as plt

content_by_years = df["release_years"].size().unstack().fillna(0)

plt.subplot(1 , 2 , figsize=[12 , 5])   
plt.plot(content_by_years.index , content_by_years["Movie"] , color = "blue")
plt.title("Movies Released per Year")
plt.xlabel("Year")
plt.ylabel("Numer Of Movies")

plt.plot(content_by_years.index , content_by_years["TV Show"] , color = "orange")
plt.title("TV Shows Released per Year")
plt.xlabel("Year")
plt.ylabel("Numer Of Movies")

plt.suptitle("Comparison of Movies and TV Shows Released per Years")

plt.tight_layout()

plt.savefig("movies_tv_show.png")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


print(df.columns)

content_by_years = df.groupby(['release_year', 'type']).size().unstack().fillna(0)


plt.figure(figsize=[12,5])

# Plot Movies

plt.plot(content_by_years.index, content_by_years.get("Movie", 0), color="blue", label="Movie")

# Plot TV Shows

plt.plot(content_by_years.index, content_by_years.get("TV Show", 0), color="orange", label="TV Show")

# Titles and labels

plt.title("Comparison of Movies and TV Shows Released per Year")
plt.xlabel("Year")
plt.ylabel("Number of Contents")
plt.legend()

plt.tight_layout()

# Save figure

plt.savefig("movies_tv_show.png")

plt.show()
```
